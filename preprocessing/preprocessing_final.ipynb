```python
import fitz

def extract_text_except_table(pdf_path, start_page, end_page):
    doc = fitz.open(pdf_path)

    if start_page < 1 or start_page > doc.page_count or end_page < start_page or end_page > doc.page_count:
      print("Invalid page range")
      return

    text_except_table = ""

    for page_num in range(start_page - 1, end_page):
        page = doc[page_num]
        blocks = page.get_text("blocks")

        for b in blocks:
            x, y, _, _ = b[:4]  # 좌표 정보

            # 예시: x 좌표가 특정 값 이상인 경우에만 텍스트 추출
            if page_num == 0:
              if x > 200 and y < 600:
                  text_except_table += b[4]
            if page_num == 1:
              if x> 100 and 100< y < 300:
                text_except_table += b[4]
    doc.close()
    return text_except_table

if __name__ == "__main__":
    for pdf_name in folder_list:
      pdf_path = folder_path + pdf_name  # PDF 파일 경로
      start_page = 1
      end_page = 2 # PDF 파일에서 텍스트 줄글이 있는 페이지만큼 수정

      extracted_text = extract_text_except_table(pdf_path, start_page, end_page)
      print(extracted_text)
      with open(folder_path + f"text_final/text_{pdf_name}.txt",'a') as file:
        file.write(extracted_text)
        file.close()
```
